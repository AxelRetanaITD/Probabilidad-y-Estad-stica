```python
# Basado en el libro estadistica para administracion pagina 12
# Vamos a crear una lista que representa la edad de 100 personas
import random
edades=[]
# Usamos un ciclo para llenar la matriz de edad
for i in range(100):
  numero_aleatorio=random.randint(18, 70)
  edades.append(numero_aleatorio)
print(edades)
```

```python
import pandas as pd
```

```python
# Creamos un DataFrame de pandas con los numeros aleatorios
df= pd.DataFrame(edades, columns=['Edad'])
df.head()
```

```python
tabla_frecuencias=df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns=['Edad','Frecuencia']
tabla_frecuencias=tabla_frecuencias.sort_values(by='Edad')
print("Tabla de Distribución de Frecuencias")
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Relativa']=tabla_frecuencias['Frecuencia']/100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Porcentual']=tabla_frecuencias['Frecuencia']/100*100
tabla_frecuencias
```

```python
tabla_frecuencias['Frec Acumulada']=tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

#Regla de sturges
Es una regla que sirve para calcular el número de clases o intervalos idoneo en los que se debe dividir un conjunto de datos.
Establece que el número de clases es igual a 1 + el logaritmo en base dos del número total de datos

```python
# Importamos la librería math de funciones aritméticas
import math
# Volvemos a usar la lista de edades
print(edades)
```

```python
# Calcular el número de intervalos utilizando la regla de Sturges
n=len(edades)
K=int(1+math.log2(100))
print("Valor de n = ",n," \nNúmero de intervalos = ",K)
```

```python
# Crear a tabla de frecuencia
tabla_frecuencia=pd.cut(df['Edad'], bins=K, include_lowest=True).value_counts().reset_index()
tabla_frecuencia.columns=['Clase','Frec']
tabla_frecuencia
```

```python
# Ordenar la tabla de frecuencia por el intervalo
tabla_frecuencia=tabla_frecuencia.sort_values(by='Clase')
tabla_frecuencia
```

```python
# Generar columnas siguientes
tabla_frecuencia['Frec Rel']=tabla_frecuencia['Frec']/n
tabla_frecuencia['Frec Porc']=tabla_frecuencia['Frec']/n*100
tabla_frecuencia['Frec Ac']=tabla_frecuencia['Frec'].cumsum()
tabla_frecuencia['Frec Rel Ac']=tabla_frecuencia['Frec Rel'].cumsum()
tabla_frecuencia['Frec Porc Ac']=tabla_frecuencia['Frec Porc'].cumsum()
# Mostrar la tabla de frecuencia
print(tabla_frecuencia)
```

```python
# Creamos un histograma basado en la tabla de frecuencia
plt.hist(df['Edad'], bins=K, alpha=0.5)
plt.title("Histograma basado en la regla de Sturges")
plt.xlabel('Edades')
plt.ylabel('Frecuencia')
```

#Practica 5: Diagrama de cajas y bigotes
Es un muy parecido al histograma y se representa con una barra en medio y un rectángulo que representa la mayor distribución de los datos, con una linea que muestra la media de los datos. Puede representar datos atípicos.


```python
edades2=[]
# Usamos un ciclo para llenar la matriz de edades 2 para realizar el diagrama de cajas y bigotes
for i in range(100):
  numero_aleatorio2=random.randint(20, 87)
  edades2.append(numero_aleatorio2)
print(edades2)
```

```python
fig, (ax1, ax2)= plt.subplots(nrows=1, ncols=2, figsize=(9,4))
bplot1=ax1.boxplot(edades,
                   vert=True, #caja vertical
                   patch_artist=True)#Llena con color
ax1.set_title("Edades Encuesta 1")
bplot2=ax2.boxplot(edades2,
                   vert=True, #caja vertical
                   patch_artist=True)#Llena con color
ax2.set_title("Edades Encuesta 2")
plt.show()
```
