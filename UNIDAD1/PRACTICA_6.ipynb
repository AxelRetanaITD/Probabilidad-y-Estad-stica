## Paso 2: Realizar un histograma de precios de los autos

```python
# Usaremos la regla de Sturge spara calcular los bins que se van a utilizar
import math
n=len(df_carro)
K=int(1+math.log2(100))
print("Valor de n = ",n," \nNúmero de intervalos = ",K)
```

```python
plt.pyplot.hist(df_carro['price'], bins=K)
plt.pyplot.title("Histograma de precios de los autos")
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Frecuencia")
```

Doy a la conclusión que los carros más baratos son los que salieron más porque son los más accesibles para las personas, a diferencia de los carros caros, que solo pueden ser comprados por cierto grupo de personas que si pueda darse ese lujo

## Paso 3: Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automovil(price) y tipo de vehículo(body style)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='body-style', y='price', data=df_carro)
plt.title("Precio vs. Tipo de vehículo")
plt.xlabel("Tipo de vehículo")
plt.ylabel("Precio")
plt.show()
```

En mi conclusión, la comparación parece mostrar que los carros de tipo hardtop son los más versátiles en cuanto a precio se refiere, a comparación de los de tipo wagon, los cuales tienen un rango de precio más cerrado, el diagrama también muesta que el costo promedio de los hatchback es el más bajo de los otros tipos.

## Paso 4: Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automovil y el tipo tracción de puertas(drive wheels)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=df_carro)
plt.title("Diagrama de caja y bigotes: Precio vs. Tipo de tracción de ruedas")
plt.xlabel("Tipo de tracción de ruedas")
plt.ylabel("Precio")
plt.show()
```

En conclusión, los carros con tracción trasera tienden a ser más caros que carros con tracción delantera o en las cuatro ruedas

## Paso 5:  Realizar un diagrama de caja y bigotes que muestre la relación estadística entre la columna del precio del automovil y la ubicación del motor (engine-location)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='engine-location', y='price', data=df_carro)
plt.title("Diagrama de caja y bigotes: Precio vs. Ubicación del motor")
plt.xlabel("Ubicación del motor")
plt.ylabel("Precio")
plt.show()
```

En conclusión: los carros con motores en la parte trasera, en promedio, son más caros que los carros con el motor en el frente
