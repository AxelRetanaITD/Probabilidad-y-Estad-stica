## Paso 2: Realizar un histograma de precios de los autos

```python
# Usaremos la regla de Sturge spara calcular los bins que se van a utilizar
import math
n=len(df_carro)
K=int(1+math.log2(100))
print("Valor de n = ",n," \nNmero de intervalos = ",K)
```

```python
plt.pyplot.hist(df_carro['price'], bins=K)
plt.pyplot.title("Histograma de precios de los autos")
plt.pyplot.xlabel("Precio")
plt.pyplot.ylabel("Frecuencia")
```

Doy a la conclusin que los carros ms baratos son los que salieron ms porque son los ms accesibles para las personas, a diferencia de los carros caros, que solo pueden ser comprados por cierto grupo de personas que si pueda darse ese lujo

## Paso 3: Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automovil(price) y tipo de vehculo(body style)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='body-style', y='price', data=df_carro)
plt.title("Precio vs. Tipo de vehculo")
plt.xlabel("Tipo de vehculo")
plt.ylabel("Precio")
plt.show()
```

En mi conclusin, la comparacin parece mostrar que los carros de tipo hardtop son los ms verstiles en cuanto a precio se refiere, a comparacin de los de tipo wagon, los cuales tienen un rango de precio ms cerrado, el diagrama tambin muesta que el costo promedio de los hatchback es el ms bajo de los otros tipos.

## Paso 4: Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automovil y el tipo traccin de puertas(drive wheels)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='drive-wheels', y='price', data=df_carro)
plt.title("Diagrama de caja y bigotes: Precio vs. Tipo de traccin de ruedas")
plt.xlabel("Tipo de traccin de ruedas")
plt.ylabel("Precio")
plt.show()
```

En conclusin, los carros con traccin trasera tienden a ser ms caros que carros con traccin delantera o en las cuatro ruedas

## Paso 5:  Realizar un diagrama de caja y bigotes que muestre la relacin estadstica entre la columna del precio del automovil y la ubicacin del motor (engine-location)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='engine-location', y='price', data=df_carro)
plt.title("Diagrama de caja y bigotes: Precio vs. Ubicacin del motor")
plt.xlabel("Ubicacin del motor")
plt.ylabel("Precio")
plt.show()
```

En conclusin: los carros con motores en la parte trasera, en promedio, son ms caros que los carros con el motor en el frente
