```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import statsmodels.api as sm
```

#Grupo 1

```python
df = pd.read_csv("grupo1.csv")
print(df.head())

print(df.describe())

plt.hist(df['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de calificaciones del grupo 1')
plt.xlabel('Calificación Final')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
X = df[['HorasEstudio']]
y = df['Calificacion']

modelo = LinearRegression()
modelo.fit(X, y)

pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print(f"Ecuación de la recta: y = {pendiente:.2f}x + {intercepto:.2f}")
```

```python
correlacion, p_valor = pearsonr(df['HorasEstudio'], df['Calificacion'])
print(f"Coeficiente de correlación de Pearson: {correlacion:.3f}")
```

```python
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df)
plt.plot(df['HorasEstudio'], modelo.predict(X), color='red', label='Línea de regresión')
plt.title('Horas de Estudio vs Calificación Final')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.show()
```

##1.-¿Explicar cuál es la interpretación del coeficiente de regresión en términos de la relación entre el tiempo de estudio y las calificaciones finales?
En el grupo 1, se observa una pendiente de -0.11, lo que indica que, en promedio, por cada hora adicional de estudio, la calificación final disminuye en 0.11 puntos. Un resultado contrario a lo esperado: que entre más horas estudiadas, mejor calificación.

##2. ¿Cómo se interpreta el coeficiente de correlación en términos de la fuerza y dirección de la relación entre las variables?
El coeficiente de correlación de Pearson es -0.057, lo cual indica una relación muy débil y negativa entre las horas de estudio y las calificaciones. En la práctica, este valor tan cercano a 0 sugiere que no existe una correlación significativa entre las variables, tal y como se presenta en la gráfica.

##3. ¿El modelo de regresión lineal es estadísticamente significativo?
El hecho de que la pendiente sea muy pequeña y el coeficiente de correlación sea cercano a 0 indica que es muy probable que el modelo no es estadísticamente significativo.

##4. ¿Qué tan confiable es el modelo de regresión para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?
No es confiable, no sigue los resultados esperados y a cambio, muestra resultados diversos y sin un patrón claro, no es posible predecir calificaciones a partir de estos datos.
