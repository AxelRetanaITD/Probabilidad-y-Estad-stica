```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import statsmodels.api as sm
```

#Grupo 1

```python
df = pd.read_csv("grupo1.csv")
print(df.head())

print(df.describe())

plt.hist(df['Calificacion'], bins=10, color='skyblue', edgecolor='black')
plt.title('Histograma de calificaciones del grupo 1')
plt.xlabel('Calificacin Final')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
X = df[['HorasEstudio']]
y = df['Calificacion']

modelo = LinearRegression()
modelo.fit(X, y)

pendiente = modelo.coef_[0]
intercepto = modelo.intercept_
print(f"Ecuacin de la recta: y = {pendiente:.2f}x + {intercepto:.2f}")
```

```python
correlacion, p_valor = pearsonr(df['HorasEstudio'], df['Calificacion'])
print(f"Coeficiente de correlacin de Pearson: {correlacion:.3f}")
```

```python
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=df)
plt.plot(df['HorasEstudio'], modelo.predict(X), color='red', label='Lnea de regresin')
plt.title('Horas de Estudio vs Calificacin Final')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacin Final')
plt.legend()
plt.grid(True)
plt.show()
```

##1.-Explicar cul es la interpretacin del coeficiente de regresin en trminos de la relacin entre el tiempo de estudio y las calificaciones finales?
En el grupo 1, se observa una pendiente de -0.11, lo que indica que, en promedio, por cada hora adicional de estudio, la calificacin final disminuye en 0.11 puntos. Un resultado contrario a lo esperado: que entre ms horas estudiadas, mejor calificacin.

##2. Cmo se interpreta el coeficiente de correlacin en trminos de la fuerza y direccin de la relacin entre las variables?
El coeficiente de correlacin de Pearson es -0.057, lo cual indica una relacin muy dbil y negativa entre las horas de estudio y las calificaciones. En la prctica, este valor tan cercano a 0 sugiere que no existe una correlacin significativa entre las variables, tal y como se presenta en la grfica.

##3. El modelo de regresin lineal es estadsticamente significativo?
El hecho de que la pendiente sea muy pequea y el coeficiente de correlacin sea cercano a 0 indica que es muy probable que el modelo no es estadsticamente significativo.

##4. Qu tan confiable es el modelo de regresin para predecir las calificaciones finales de los estudiantes basado en el tiempo de estudio?
No es confiable, no sigue los resultados esperados y a cambio, muestra resultados diversos y sin un patrn claro, no es posible predecir calificaciones a partir de estos datos.
